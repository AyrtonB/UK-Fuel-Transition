### Helper Functions

```python
def hide_spines(ax, positions=['top', 'right']):
    """
    Pass a matplotlib axis and list of positions with spines to be removed
    
    args:
        ax:          Matplotlib axis object
        positions:   Python list e.g. ['top', 'bottom']
    """
    assert isinstance(positions, list), 'Position must be passed as a list '
    
    for position in positions:
        ax.spines[position].set_visible(False)
        
def try_accept(exception, default=None):
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            try:
                return func(*args, **kwargs)
            except:
                return exception
        return wrapper
    return decorator
```

### Preparing Camel Curve

We will focus on May, June & July (5, 6, 7)

```python
df_solar_focus = df_EI.loc[df_EI.index.month.isin([5, 6, 7]), ['solar']]
df_demand_focus = df_EI.loc[df_EI.index.month.isin([5, 6, 7]), ['demand']]

df_solar_focus.head()
```

```python
def create_df_dt_rng(start_date, end_date, freq='30T', tz='Europe/London', dt_str_template='%Y-%m-%d'):
    ## Creating localised datetime index
    s_dt_rng = pd.date_range(start_date, end_date, freq=freq, tz=tz)
    s_dt_SP_count = pd.Series(0, index=s_dt_rng).resample('D').count()

    ## Creating SP column
    SPs = []
    for num_SPs in list(s_dt_SP_count):
        SPs += list(range(1, num_SPs+1))

    ## Creating datetime dataframe
    df_dt_rng = pd.DataFrame(index=s_dt_rng)
    df_dt_rng.index.name = 'local_datetime'

    ## Adding query call cols
    df_dt_rng['SP'] = SPs
    df_dt_rng['date'] = df_dt_rng.index.strftime(dt_str_template)#.str.replace(' ', '%20')

    return df_dt_rng

df_dt_rng = create_df_dt_rng(df_solar_focus.index.min(), df_solar_focus.index.max())
df_dt_rng.index.freq = None

df_dt_rng.head()
```

```python
ts_2_SP = df_dt_rng['SP'].to_dict()
SP_2_ts = df_dt_rng.reset_index().set_index('SP')['local_datetime'].to_dict()

df_solar_focus['SP'] = df_solar_focus.index.map(ts_2_SP)
df_demand_focus['SP'] = df_demand_focus.index.map(ts_2_SP)

df_solar_focus.head()
```

```python
df_solar_focus['year'] = df_solar_focus.index.year
df_solar_focus['doy'] = df_solar_focus.index.dayofyear

years = df_solar_focus['year'].unique()
max_solar_time_dict = dict()

for year in years:
    max_solar_time_dict[year] = []
    df_solar_focus_year = df_solar_focus[df_solar_focus['year']==year]
    
    for doy in df_solar_focus['doy'].unique():
        df_solar_focus_year_day = df_solar_focus_year[df_solar_focus_year['doy']==doy]
        
        if df_solar_focus_year_day.shape[0] > 0:
            max_solar_time = df_solar_focus_year_day['solar'].idxmax()
            max_solar_time_dict[year] += [max_solar_time.hour+max_solar_time.minute/30]
            
for year in years[2:]:
    sns.distplot(max_solar_time_dict[year], kde_kws=dict(bw=0.75), hist=False)
```

### Deriving Cardinal Points

```python
fig, ax = plt.subplots(dpi=250)

for idx, year in enumerate(years):
    s_dem_min_sol_year = df_dem_min_sol[year]
    s_dem_min_sol_year.plot(color=palette[idx], ax=ax)
    
hide_spines(ax)
```

```python
def locate_year_cardinals(df_year):
    n = 3 # number of points to be checked before and after 

    # Find local peaks
    df_year['min'] = df_year.iloc[argrelextrema(df_year.dem_min_sol.values, np.less_equal, order=n)[0]]['dem_min_sol']
    df_year['max'] = df_year.iloc[argrelextrema(df_year.dem_min_sol.values, np.greater_equal, order=n)[0]]['dem_min_sol']

    s_year_cardinals = df_year['min'].dropna().append(df_year['max'].dropna())

    return s_year_cardinals

year = 2018

df_year = pd.DataFrame({'dem_min_sol':df_dem_min_sol[year]})
s_year_cardinals = locate_year_cardinals(df_year)

s_year_cardinals
```

<br>

Determining the cardinal times and values for each year

```python
points = ['a', 'b', 'c', 'd', 'e', 'f']

idx_arrays = np.array([sorted(2*points), ['time', 'dem_min_sol']*6])
new_multi_idx = pd.MultiIndex.from_arrays(idx_arrays)

df_cardinals = pd.DataFrame(columns=new_multi_idx, index=years)
df_cardinals.columns.names = ['cardinal', 'value/time']

for year in years:
    df_year = pd.DataFrame({'dem_min_sol':df_dem_min_sol[year]})
    s_year_cardinals = locate_year_cardinals(df_year)

    list_of_cardinal_tuples = list(zip(s_year_cardinals.index, s_year_cardinals))
    formated_cardinal_list = list(itertools.chain.from_iterable(list_of_cardinal_tuples))

    df_cardinals.loc[year] = formated_cardinal_list

df_cardinals
```

```python
def create_solar_means(df_solar_focus_daily_wide, SP_2_ts, year=2018):
    s_SP_solar_means = df_solar_focus_daily_wide[year]
    s_SP_solar_means.index = s_SP_solar_means.index.map(SP_2_ts)
    s_SP_solar_means.index = s_SP_solar_means.index.map(lambda ts: ts.replace(year=2019, month=6, day=1))

    return s_SP_solar_means

s_SP_solar_means = create_solar_means(df_solar_focus_daily_wide, SP_2_ts, year=2018)
s_SP_solar_means.head()
```

```python
def plot_duck_curve_ax(ax, years, df_cardinals_times, df_cardinals_values, palette):
    for idx, year in enumerate(years):
        times, values = df_cardinals_times.loc[year].values, df_cardinals_values.loc[year].values
        
        if isinstance(df_cardinals_times.iloc[0, 0], datetime.time):
            times = [datetime.datetime.combine(datetime.date(2019, 6, 1), time) for time in times]
        
        ax.scatter(times, values, color=palette[idx], s=10)

        s_dem_min_sol_year = df_dem_min_sol[year].copy()
        s_dem_min_sol_year_index = [datetime.datetime.combine(datetime.date(2019, 6, 1), time) for time in s_dem_min_sol_year.index]
        
        ax.plot(s_dem_min_sol_year_index, s_dem_min_sol_year, color=palette[idx], lw=1, alpha=0.75, label=year)

    hide_spines(ax)

    ax.set_xlabel('')
    ax.set_ylabel('Demand Minus Solar Output (GW)')
    
    return
    
grid = plt.GridSpec(4, 1, wspace=0.4, hspace=0.3)

fig = plt.figure(dpi=250)
ax1 = plt.subplot(grid[0, 0])
ax2 = plt.subplot(grid[1:, 0])

# Solar profile axis
s_SP_solar_means_2018 = create_solar_means(df_solar_focus_daily_wide.copy(), SP_2_ts, year=2018)
s_SP_solar_means_2019 = create_solar_means(df_solar_focus_daily_wide.copy(), SP_2_ts, year=2019)

ax1.fill_between(s_SP_solar_means_2018.index, s_SP_solar_means_2018, color='gold', label='2018')
ax1.plot(s_SP_solar_means_2019.index, s_SP_solar_means_2018, color='k', alpha=0.2, label=None)

ax1.fill_between(s_SP_solar_means_2019.index, s_SP_solar_means_2019, color='orange', alpha=1, label='2019')
ax1.plot(s_SP_solar_means_2019.index, s_SP_solar_means_2019, color='k', alpha=0.2, label=None)

hide_spines(ax1, positions=['top', 'bottom', 'right'])
ax1.set_xticks([], [])
ax1.set_xlabel('')
#ax1.set_yticks([], [])
ax1.set_ylabel('Solar (GW)')
ax1.set_xlim(pd.Timestamp('2019-06-01 00:00'), pd.Timestamp('2019-06-01 23:30'))

ax1.legend(frameon=False)

## Duck curve axis
plot_duck_curve_ax(ax2, years, df_cardinals_times[['f', 'b', 'e']], df_cardinals_values[['f', 'b', 'e']], palette)

ax2.set_xlim(pd.Timestamp('2019-06-01 00:00'), pd.Timestamp('2019-06-01 23:30'))
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%H'))

ax2.set_xlabel('Time (BST)')
#ax2.set_ylim(0)

## Figure parameters
fig.suptitle("The UK Solar 'Camel Curve' - May to July")
```

```python
def time_2_timestamp(time, date=datetime.date(2019, 6, 1)):
    datetime_val = datetime.datetime.combine(date, time)
    timestamp_val = pd.to_datetime(datetime_val)
    
    return timestamp_val

#df_cardinals_times.apply(pd.to_datetime, axis=1)
```

```python
for col in df_cardinals_times.columns:
    df_cardinals_times.loc[:, col] = df_cardinals_times.loc[:, col].apply(time_2_timestamp)
    
df_cardinals_times.head(1)
```

```python
s_rampdown_time = df_cardinals_times['e'] - df_cardinals_times['b'] 
s_rampup_time = df_cardinals_times['f'] - df_cardinals_times['b'] 

s_rampdown_time_hrs = s_rampdown_time.dt.total_seconds()/60**2
s_rampup_time_hrs = s_rampup_time.dt.total_seconds()/60**2
```

```python
s_rampdown = df_cardinals_values['e'] - df_cardinals_values['b'] 
s_rampup = df_cardinals_values['f'] - df_cardinals_values['b'] 
```

```python
s_rampdown_rate = s_rampdown / s_rampdown_time_hrs
s_rampup_rate =  s_rampup / s_rampup_time_hrs

s_rampdown_rate
```

```python
fig, axs = plt.subplots(dpi=750, ncols=2, figsize=(10, 5))

## Ramp-up
sns.regplot(s_annual_daily_output.index.values, -s_rampdown_rate, ax=axs[0], color='darkorange')

hide_spines(axs[0])
axs[0].set_xlabel('')
axs[0].set_ylabel('Average Midday Ramp Down (GW/hr)')

## Ramp-up
sns.regplot(s_annual_daily_output.index.values, s_rampup_rate, ax=axs[1], color='darkorange')

hide_spines(axs[1])
axs[1].set_xlabel('')
axs[1].set_ylabel('Average Evening Ramp Up (GW/hr)')

fig.savefig('img/Ramping.png')
```
