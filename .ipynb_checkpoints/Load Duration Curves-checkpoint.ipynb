```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

#### Reading in Data

```python
df_EI = pd.read_csv('data/EI.csv', parse_dates=['UTC']).rename(columns={'UTC':'datetime'})
df_EI.head()
```

<br>

#### Demand Load Curve

```python
s_demand = df_EI.set_index('datetime')['demand']
s_demand.head()
```

```python
def sort_series(s_demand, year):
    s_sorted_demand = (s_demand
                       .copy()
                       [str(year)]
                       .sort_values(ascending=False)
                       .reset_index(drop=True)
                      )
    return s_sorted_demand

years = list(range(2009, 2018))

### ******** Use enumerate and a colormap

plt.figure(dpi=150)
for year in years:
    s_sorted_demand = sort_series(s_demand, year)
    plt.plot(s_sorted_demand.index, s_sorted_demand, label=f'{year}')
    
plt.legend(frameon=False)
```

#### Plotting Individual Fuel Load Curves

```python
fuels = ['biomass', 'coal', 'gas', 'hydro', 'nuclear', 'solar', 'wind', 'pumped_storage']
df_EI.columns
```

```python
def plot_fuel(fuel, df_energy):
    plt.figure()
    for year in years:
        s_fuel = df_EI.set_index('datetime')[fuel]
        s_sorted_fuel = sort_series(s_fuel, year)
        plt.plot(s_sorted_fuel.index, s_sorted_fuel, label=f'{year}')

    plt.legend(frameon=False)
    plt.title(f'{fuel.capitalize()} Load Curve')

for fuel in fuels:
    plot_fuel(fuel, df_EI)
```
