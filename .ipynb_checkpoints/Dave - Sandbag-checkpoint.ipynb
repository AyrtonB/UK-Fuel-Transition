### Helper Functions

```python
def df_country_long_2_wide(df_country, country):
    df_country['year'] = df_country.index.year.astype(str)
    df_country['doy'] = df_country.index.dayofyear

    df_country_wide = df_country.pivot(index='year', columns='doy', values=country)
    
    return df_country_wide

def plot_wide_country(df_country_wide, country, ax, bkg_c='white', cmap='viridis', v_min_max='df', yticklabels=list(range(2015,2019+1))):
    v_min_max_dict = {
        v_min_max : (1, 1),
        'df' : (None, None),
        '0-1' : (0, 1),
        '0-100' : (0, 100),
    }
    
    vmin, vmax = v_min_max_dict[v_min_max]
    
    ## Creating heatmap
    sns.heatmap(df_country_wide*100, cmap=cmap, ax=ax, mask=df_country_wide.isnull(), vmin=vmin, vmax=vmax,
                cbar_kws={'use_gridspec':False, 'shrink':0.65, 'location':'top', 
                          'anchor':(1, -0.4), 'format':'%.0f%%', 'aspect':10})
    
    cbar = ax.collections[0].colorbar
    cbar.ax.tick_params(labelsize=16, top=False, pad=0.015)
    #cbar.outline.set_color('k')
    #cbar.outline.set_linewidth(0.001)
            
    ## Setting main and axis titles
    #ax.set_title(f"{country.replace('_', ' ').title()}", y=1.2, x=0.25, size=25)
    ax.set_xlabel('')
    ax.set_ylabel('')
    
    ## Formatting ticks
    month_starts, month_names = create_month_starts_names()
    ax.set_yticklabels(yticklabels, fontsize=10)

    ax.hlines(list(range(1, 11)), xmin=0, xmax=365, color=[bkg_c])
    ax.tick_params(left=False, bottom=False, labelbottom=False, pad=0.5)
        
    return ax

def plot_wide_country_side(df_country_wide, country, ax, bkg_c='white', cmap='viridis', v_min_max='df', yticklabels=list(range(2015,2019+1))):
    v_min_max_dict = {
        v_min_max : (1, 1),
        'df' : (None, None),
        '0-1' : (0, 1),
        '0-100' : (0, 100),
    }
    
    vmin, vmax = v_min_max_dict[v_min_max]
    
    ## Creating heatmap
    sns.heatmap(df_country_wide*100, cmap=cmap, ax=ax, mask=df_country_wide.isnull(), vmin=vmin, vmax=vmax,
                cbar_kws={'use_gridspec':False, 'shrink':1, 'pad':0.025,
                          'format':'%.0f%%', 'aspect':10})
    
    cbar = ax.collections[0].colorbar
    cbar.ax.tick_params(labelsize=16, top=False, pad=0.015)
    #cbar.outline.set_color('k')
    #cbar.outline.set_linewidth(0.001)
            
    ## Setting main and axis titles
    #ax.set_title(f"{country.replace('_', ' ').title()}", y=1.2, x=0.25, size=25)
    ax.set_xlabel('')
    ax.set_ylabel('')
    
    ## Formatting ticks
    month_starts, month_names = create_month_starts_names()
    ax.set_yticklabels(yticklabels, fontsize=10)

    ax.hlines(list(range(1, 11)), xmin=0, xmax=365, color=[bkg_c])
    ax.tick_params(left=False, bottom=False, labelbottom=False, pad=0.5)
        
    return ax
                 
def make_colormap(seq):
    """Return a LinearSegmentedColormap
    seq: a sequence of floats and RGB-tuples. The floats should be increasing
    and in the interval (0,1).
    """
    seq = [(None,) * 3, 0.0] + list(seq) + [1.0, (None,) * 3]
    cdict = {'red': [], 'green': [], 'blue': []}
    for i, item in enumerate(seq):
        if isinstance(item, float):
            r1, g1, b1 = seq[i - 1]
            r2, g2, b2 = seq[i + 1]
            cdict['red'].append([item, r1, r2])
            cdict['green'].append([item, g1, g2])
            cdict['blue'].append([item, b1, b2])
    return mcolors.LinearSegmentedColormap('CustomMap', cdict)
                 
def create_month_starts_names():
    s_doy = pd.Series(index=pd.date_range('2018', '2019', freq='30T')).resample('D').mean()
    s_doy.loc[:] = s_doy.index.dayofyear

    month_starts = s_doy.groupby(s_doy.index.month).min() + 15
    month_names = s_doy.index.month_name().unique()

    return month_starts, month_names



c = mcolors.ColorConverter().to_rgb
rgb_2_color = lambda rgb_list: [elem/255 for elem in rgb_list]

fuel_cmaps = {
     'green_grey' : [rgb_2_color([10, 204, 80]), rgb_2_color([10, 204, 80]), 0.0001, rgb_2_color([10, 204, 80]), rgb_2_color([250, 245, 135]), 0.05, c('white'), c('black')], # orig green -> rgb_2_color([0, 204, 68]
     'grey' : [c('white'), c('black')],
     'green' : [rgb_2_color([0, 204, 68]), c('black')]}
                     
month_starts, month_names = create_month_starts_names()
month_name_2_DoY = dict(zip(month_names, month_starts))
```

### Flag Plotting Helper Functions

```python
def get_flag(name):
    path = f'sandbag/{name.title()}.png'
    im = plt.imread(path)
    return im

def offset_image(coord, name, ax):
    img = get_flag(name)
    im = OffsetImage(img, zoom=0.72)
    im.image.axes = ax

    ab = AnnotationBbox(im, (coord, 0),  xybox=(0., -16.), frameon=False,
                        xycoords='data',  boxcoords="offset points", pad=0)

    ax.add_artist(ab)
```

### Missing Data

```python
dt_rng = pd.date_range(df.index[0], df.index[-1], freq='D')
df = df.reindex(dt_rng)

df_resampled_nulls = df.isnull().astype(int).resample('7D').mean()
df_resampled_nulls = df_resampled_nulls['2018-11':]
cols_with_nulls = df_resampled_nulls.columns[df_resampled_nulls.sum()!=0]

df_resampled_nulls = df_resampled_nulls[cols_with_nulls]
df_resampled_nulls.index = df_resampled_nulls.index.strftime('%Y-%m')

sns.heatmap(df_resampled_nulls, cmap='coolwarm', mask=df_resampled_nulls==0, vmin=0, vmax=1)
```

```python
num_missing = df.isnull().sum(axis=1)
df_missing = df[num_missing>0]

deltas = df_missing.index - pd.Series(df_missing.index).shift(1)
delta_days = deltas.apply(lambda x: x.days)
```

```python
delta_days.head()
```

```python
group = 0
groups = []

for idx, days_delta in delta_days.iteritems():
    if days_delta != 1:
        group += 1
        
    groups += [group]
    
s_groups = pd.Series(groups)
s_group_counts = s_groups.groupby(s_groups).count()
s_batch_group_counts = s_group_counts[s_group_counts!=1]

s_groups.index = df_missing.index
start_date_idxs = s_groups[~s_groups.duplicated()].index

df_missing_date_groups = pd.DataFrame(data={'days_in_group': s_batch_group_counts.values}, index=start_date_idxs)
df_missing_date_groups = df_missing_date_groups.style.set_caption('')
df_missing_date_groups.index.name = 'Group Start Date'

print(f'There are {s_batch_group_counts.shape[0]} batches which have more than 1 missing value')
print(f'{(s_group_counts==1).sum()} datetimes contained missing values which weren\'t part of a larger group\n')

df_missing_date_groups
```

### Plotting Single Country

```python
def plot_country_clean_side(df_country_wide, country, fuel_cmaps):    
    logo = image.imread('img/sandbag/sandbaglogo_simple_1000.png')

    fig, ax = plt.subplots(figsize=(14, 10), dpi=500)

    ## Creating wide df for fuel
    df_country = df.loc[:, [country]]
    df_country_wide = df_country_long_2_wide(df_country, country)

    ## Removing extra days caused by leap years
    cols_2_drop = list(df_country_wide.columns[df_country_wide.isnull().sum()>1])
    df_country_wide = df_country_wide.drop(columns=cols_2_drop)

    ## Plotting    
    ax = plot_wide_country_side(df_country_wide, country, ax, cmap=make_colormap(fuel_cmaps['green_grey']), v_min_max='0-100')
    ax.set_yticklabels(list(range(2015, 2019+1)), va='center', size=25)

    fig.suptitle(country, y=0.95, x=0.205, fontsize=50)
    
    ## Annotating
    source_text = 'Source: Sandbag, using ENTSO-E data; viz by Ayrton'
    ax.text(90, 5.1, source_text, size=12, ha='center', va='center')
    
    newax = fig.add_axes([0.45, 0.045, 0.3, 0.3], zorder=1)
    newax.imshow(logo)
    newax.axis('off')
    
    fig.savefig(f'img/sandbag/{country}_sidebar.png')

for country in ['France']:
    plot_country_clean_side(df_country_wide, country, fuel_cmaps)
```

```python
def plot_country_clean_top(df_country_wide, country, fuel_cmaps):
    logo = image.imread('img/sandbag/sandbaglogo_simple_1000.png')
    
    fig, ax = plt.subplots(figsize=(10, 10), dpi=500)

    ## Creating wide df for fuel
    df_country = df.loc[:, [country]]
    df_country_wide = df_country_long_2_wide(df_country, country)

    ## Removing extra days caused by leap years
    cols_2_drop = list(df_country_wide.columns[df_country_wide.isnull().sum()>1])
    df_country_wide = df_country_wide.drop(columns=cols_2_drop)

    ## Plotting    
    ax = plot_wide_country(df_country_wide, country, ax, cmap=make_colormap(fuel_cmaps['green_grey']), v_min_max='0-100')
    ax.set_yticklabels(list(range(2015, 2019+1)), va='center', size=25)

    fig.suptitle(country, y=0.8, x=0.237, fontsize=50)
    
    ## Annotating
    source_text = 'Source: Sandbag, using ENTSO-E data; viz by Ayrton'
    ax.text(90, 5.1, source_text, size=10.5, ha='center', va='center')
    
    newax = fig.add_axes([0.53, 0.005, 0.35, 0.35], zorder=1)
    newax.imshow(logo)
    newax.axis('off')
    
    fig.savefig(f'img/sandbag/{country}_topbar.png')

for country in ['France']:
    plot_country_clean_top(df_country_wide, country, fuel_cmaps)
```
